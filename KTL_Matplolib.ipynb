  - add_subplot() 함수와 add_axes() 함수로 출력한 그래프에는 차이가 있다.
  - add_subplot() 함수는 전체 figure를 배열의 그리드로 구분한다. 반면 add_axes는 figure 좌측 하단을 0으로 하여 메소드 인수로 x, y 기준점을 잡고 그 기준점에서 너비와 높이를 계산하여 그래프를 그린다.(예, add_axes([1,2,3,4] 는 기준점에서 x축으로 1, y축으로 2만큼 떨어진곳을 그림의 좌측하단 기준점으로 해서, 가로축의 폭은 4, 세로축의 폭은 3인 그래프가 그려진다.)

```python
fig=plt.figure()
ax1=fig.add_subplot(3,4,1)
ax2=fig.add_subplot(3,4,3)
ax3=fig.add_axes([0.5,0.1,0.4,0.3])
ax4=fig.add_axes([0.8,0.3,0.2,0.5])
ax1.set_title('ax1')
ax2.set_title('ax2')
ax3.set_title('ax3')
ax4.set_title('ax4')
plt.show()
```

```python
fig, ax=plt.subplots(2,1, figsize=(5,5))
x=np.linspace(0,10,12)
y1=np.random.randint(10, size=12)
y2=np.random.randint(10, size=12)
y3=np.arange(12)
y4=np.random.randint(10, size=12)
lab=['line a','line b','line c','line d']
ax[0].plot(x,y1,'g-')
ax[0].plot(x,y2,'--')
ax[1].plot(x,y3,'r.')
ax[1].plot(x,y4,'b:')
ax[0].legend(labels=lab[:2], loc='upper right')
ax[1].legend(labels=lab[2:], loc=2)
plt.show()
```

    - Matplotlib은 로그와 같은 비선형 척도를 지원합니다.
      - Logit 함수는 log-odds function으로도 알려져있다. 이 함수는 0에서 1까지의 확률값과 -∞에서 ∞ 사이의 확률값을 표현해주는 함수이다. X축이 아니라 Y축에서 0과 1 사이의 값을 제한하는 시그모이드 함수에 대한 역함수이다.
      - 'symlog'는 Symmetrical Log Scale이다. 축의 원점을 기준으로 양, 음의 방향이 대칭적인 로그 스케일로 표시된다.

  - 틱과 틱커 (Ticks and tickers)
    - 각 축에는 "틱(ticks)"이라는 작은 표시가 있다. 정확히 말하자면, "틱"은 표시(예. (-1, 0, 1))의 위치"이며, 틱 선은 그 위치에 그려지는 작은 선이다. 또한 "틱 라벨"은 틱 선 옆에 그려지는 라벨이며, "틱커"는 틱의 위치를 결정하는 객체 이다. 기본적인 틱커는 ~5 에서 8 틱을 위치시키는데 꽤 잘 작동합니다. 즉, 틱 서로간에 적당한 거리를 표현합니다.
    - 하지만, 가끔은 좀 더 이를 제어할 필요가 있습니다 (예. 위의 로짓 그래프에서는 너무 많은 틱 라벨이 있습니다). 다행히도 matplotlib은 틱을 완전히 제어하는 방법을 제공합니다. 심지어 보조 눈금(minor tick)을 활성화 할 수도 있다.


```python
x=np.linspace(-2,2,100)
plt.figure(figsize=(15,10))
plt.subplot(131)
plt.plot(x, x**3)
plt.grid(True)
plt.title('default ticks')

ax=plt.subplot(132)
plt.plot(x,x**3)
ax.xaxis.set_ticks(np.arange(-2,2,1))
plt.grid(True)
plt.title('manual ticks on the x-axis')

ax1=plt.subplot(133)
plt.plot(x,x**3)
plt.minorticks_on()
ax1.xaxis.set_ticks([-2,0,1,2])
ax1.yaxis.set_ticks(np.arange(-5,5,1))
ax1.yaxis.set_ticklabels(['min',-4,-3,-2,-1,0,1,2,3,'max'])
plt.title('manual ticks and tick labels\n(plus minor ticks) on the y-axis')
plt.grid(True)
plt.show()
```

### **기타 그래프**

  - 막대그래프
    - bar() 와 barh() 함수를 사용해 막대그래프를 만들 수 있다.
    - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar
    - https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.barh.html#matplotlib.pyplot.barh

```python
girl_weight=(30,25,29,27,22,35)
boy_weight=(33,37,28,25,20,23)
girl_err=(3,2,1,2,4,1)
boy_err=(1,2,3,2,3,1)
num=np.arange(6)
width=0.6
bar1=plt.bar(num, girl_weight, width, yerr=girl_err)
bar2=plt.bar(num, boy_weight, width, bottom=girl_weight, yerr=boy_err)
plt.title('sample weight of girls and boys')
plt.ylabel('student weight', fontsize=14)
plt.xticks(num, ('s1','s2','s3','s4','s5','s6'))
plt.yticks(np.arange(0,90,5))
plt.legend((bar1, bar2),('girl','boy'))
plt.show()
```

    - pie() 함수를 이용하여 원그래프를 생성한다.
    - 기본 원그래프 외의 추가기능
      - 슬라이스 라벨
      - 백분율의 자동 라벨링
      - explode 로 슬라이스를 오프셋
      - 그림자 설정
      - 쐐기를 그리는 시작 각도 조정

```python
lbl='apple','grape','pear','lemon'
size=[35,30,25,10]
epd=(0,0.1,0,0)
fig, ax=plt.subplots()
ax.pie(size, textprops={'fontsize':14}, explode=epd, labels=lbl, autopct='%1.1f%%',
       shadow=True, startangle=90)
ax.axis('equal')
plt.show()
```

  - 3차원 그래프
    - 3차원 그래프를 그리는것은 꽤 간단하다. 우선 "3d" 투영을 등록하는 Axes3D를 임포트 한다. 그리고 나서 projection 속성을 "3d"로 설정된 부분 그래프를 생성한다. 그러면 Axes3DSubplot 이라는 객체가 반환되는데, 이 객체의 plot_surface 메서드를 호출하면 x, y, z 좌표를 포함한 추가적이나 속성을 지정할 수 있다.

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib
x=np.linspace(-5,5,50)
y=np.linspace(-5,5,50)
X, Y=np.meshgrid(x, y)
R=np.sqrt(X**2+Y**2)
Z=np.sin(R)
figure=plt.figure(figsize=(12,5))
subplot3d=plt.subplot(111, projection='3d')
surface=subplot3d.plot_surface(X,Y,Z,rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0.1)
plt.show()
```
